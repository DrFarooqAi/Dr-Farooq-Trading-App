### Convert DataFrame to Excel

This section demonstrates how to create a pandas DataFrame and then save it to an Excel (.xlsx) file. You can replace the sample DataFrame with your actual DataFrame and customize the output file name.

```python
import pandas as pd

# --- Create a sample DataFrame (replace with your actual DataFrame) ---
data = {
    'Column1': [1, 2, 3, 4, 5],
    'Column2': ['A', 'B', 'C', 'D', 'E'],
    'Column3': [10.1, 20.2, 30.3, 40.4, 50.5]
}
sample_df = pd.DataFrame(data)

print("Sample DataFrame:")
display(sample_df.head())
```

```python
# --- Specify the output Excel file name ---
output_excel_filename = 'my_output_data.xlsx'

# --- Save the DataFrame to an Excel file ---
sample_df.to_excel(output_excel_filename, index=False) # Set index=False if you don't want to write the DataFrame index to the Excel file

print(f"DataFrame successfully saved to '{output_excel_filename}'")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# ========== 1. Load your original 100-patient base (from the CSV you already have) ==========
# For this example, I recreate the 100‑patient structure from your earlier data.
# In practice, you would read your actual CSV:
# df_base = pd.read_csv('your_100_patients.csv')

# Creating 100 base patients (simplified – you can replace with your full data)
np.random.seed(42)
random.seed(42)

departments = {
    'Orthopedic': ['John Smith', 'Emily Jones', 'Michael Brown'],
    'Cardiac': ['Sarah Wilson', 'David Martinez', 'Lisa Garcia'],
    'ER': ['Robert Taylor', 'Patricia Lee', 'James Anderson'],
    'Derma': ['Jennifer Thomas', 'Charles White', 'Susan Harris'],
    'Neurology': ['Daniel Moore', 'Nancy Jackson', 'Matthew Martin'],
    'Pediatrics': ['Laura Robinson', 'Kevin Lewis', 'Amy Walker'],
    'Gynecology': ['George Hall', 'Kimberly Allen', 'Edward Young'],
    'Urology': ['Brian King', 'Steven Adams', 'Deborah Scott'],
    'Gastroenterology': ['Margaret Baker', 'Paul Nelson', 'Karen Carter'],
    'Oncology': ['Jason Mitchell', 'Rebecca Perez', 'Eric Roberts']
}

all_doctors = [doc for dept in departments.values() for doc in dept]
dept_names = list(departments.keys())

# Generate 100 base patients (matching your original fields)
first_names = ['Christina','Ryan','Linda','Anna','Juan','Miguel','Tyler','Lisa','David','Teresa',
               'Jonathan','Douglas','Anthony','Kimberly','Travis','Amanda','Olivia','James','Sophia',
               'Benjamin','Mia','Elijah','Isabella','Lucas','Amelia','Mason','Charlotte','Ethan','Ava',
               'Alexander','Sofia','Daniel','Madison','Joseph','Abigail','Samuel','Elizabeth','David',
               'Grace','Andrew','Chloe','Matthew','Victoria','Jack','Zoe','Nathan','Hannah','Dylan',
               'Leah','Gabriel','Audrey','Isaac','Lily','Carter','Penelope','Lincoln','Stella','Julian',
               'Nora','Owen','Mila','Eli','Aria','Ezra','Scarlett','Leo','Ellie','Hudson','Violet',
               'Asher','Luna','Grayson','Hazel','Wyatt','Aurora','Miles','Nova','Caleb','Isla','Elias',
               'Ava','Mateo','Layla','Cameron','Bella','Nolan','Zara','Easton','Clara','Christian',
               'Anna','Julian','Maya','Connor','Ella','Isaiah','Madelyn','Josiah','Lillian', # 97 elements, adding 3 more
               'William', 'Harper', 'Evelyn'] # Added 3 names
last_names = ['Martinez','Richmond','Wheeler','Phillips','Patel','Fields','Rosario','Johnson','Sanchez',
              'Marks','Fernandez','Crawford','Fisher','Vargas','Walker','Ortiz','Smith','Johnson','Williams',
              'Brown','Jones','Garcia','Miller','Davis','Rodriguez','Martinez','Hernandez','Lopez','Gonzalez',
              'Wilson','Anderson','Thomas','Taylor','Moore','Jackson','White','Harris','Martin','Thompson',
              'Martinez','Robinson','Clark','Rodriguez','Lewis','Young','Hall','Allen','King','Wright',
              'Scott','Green','Adams','Baker','Nelson','Carter','Mitchell','Perez','Roberts','Hill',
              'Campbell','Rivera','Foster','Diaz','Simmons','Foster','Price','Barnes','Cook','Morgan',
              'Cooper','Reed','Ward','Stewart','Sanders','Howard','Rogers','Russell','Griffin','Hayes',
              'Price','Butler','Bryant','Russell','Watson','Sanders','Reed','Powell','Perry','Morgan',
              'Bell','Murphy','Bailey','Simmons','Foster','Cox','Rivera','Patterson','Gray','Price', # 99 elements, adding 1 more
              'Kelly'] # Added 1 name
genders = ['Female','Male']*50
blood_types = ['AB-','A-','A-','O-','AB-','B-','O-','B-','O+','O+','O+','AB+','O+','B+','A-','A+']*7
conditions = ['Arthritis','Cancer','Arthritis','Obesity','Diabetes','Diabetes','Cancer','Diabetes','Arthritis',
              'Cancer','Arthritis','Arthritis','Asthma','Obesity','Arthritis','Cancer']*7
admission_dates = pd.date_range('2018-01-01','2023-12-31', periods=100)
billing = np.random.uniform(1000,45000,100)
room_numbers = [random.randint(50,500) for _ in range(100)]
admission_types = ['Urgent','Emergency','Elective','Emergency','Urgent','Elective','Emergency','Emergency',
                   'Emergency','Emergency','Emergency','Elective','Urgent','Elective','Emergency','Emergency']*7
discharge_dates = [admission_dates[i] + timedelta(days=random.randint(1,30)) for i in range(100)]
medications = ['Paracetamol','Aspirin','Lipitor','Ibuprofen','Paracetamol','Aspirin','Penicillin','Ibuprofen',
               'Aspirin','Penicillin','Lipitor','Paracetamol','Aspirin','Ibuprofen','Penicillin','Paracetamol']*7
test_results = ['Normal','Abnormal','Abnormal','Normal','Abnormal','Normal','Abnormal','Abnormal',
                'Normal','Abnormal','Normal','Abnormal','Normal','Abnormal','Normal','Abnormal']*7

base_patients = []
for i in range(100):
    dept = random.choice(dept_names)
    doctor = random.choice(departments[dept])
    base_patients.append({
        'Patient Name': f"{first_names[i]} {last_names[i]}",
        'Age': random.randint(18,90),
        'Gender': genders[i % len(genders)],
        'Blood Type': blood_types[i % len(blood_types)],
        'Medical Condition': conditions[i % len(conditions)],
        'Date of Admission': admission_dates[i].strftime('%Y-%m-%d'),
        'Doctor': doctor,
        'Department': dept,
        'Insurance Provider': random.choice(['Aetna','Blue Cross','Medicare','Cigna','UnitedHealthcare']),
        'Billing Amount': round(billing[i],2),
        'Room Number': room_numbers[i] if i >= 90 else '',  # last 10 = inpatient
        'Admission Type': admission_types[i % len(admission_types)],
        'Discharge Date': discharge_dates[i].strftime('%Y-%m-%d'),
        'Medication': medications[i % len(medications)],
        'Test Results': test_results[i % len(test_results)]
    })

# Mark patient type: first 90 = Outpatient, last 10 = Inpatient
for i, p in enumerate(base_patients):
    p['Patient Type'] = 'Inpatient' if i >= 90 else 'Outpatient'
    if p['Patient Type'] == 'Outpatient':
        p['Room Number'] = ''   # no room for outpatients

# ========== 2. Expand Outpatients into multiple OPD service rows ==========
radiology_services = ['MRI Brain', 'CT Chest', 'XRAY Foot', 'XRAY Hands', 'CT Abdomen', 'MRI Spine', 'XRAY Knee']
laboratory_tests = ['CBC', 'LFT', 'KFT', 'Lipid Profile', 'Blood Glucose', 'HbA1c', 'Thyroid Panel']

expanded_rows = []
for p in base_patients:
    if p['Patient Type'] == 'Inpatient':
        # Inpatient: single row, no radiology/lab
        row = p.copy()
        row['Radiology'] = ''
        row['Laboratory'] = ''
        row['Admission Type'] = p['Admission Type']  # keep original
        expanded_rows.append(row)
    else:
        # Outpatient: generate 4 rows (2 radiology + 2 lab)
        # Keep Admission Type empty
        for _ in range(2):
            row = p.copy()
            row['Admission Type'] = ''   # empty for OPD
            row['Radiology'] = random.choice(radiology_services)
            row['Laboratory'] = ''
            row['Discharge Date'] = (datetime.strptime(p['Date of Admission'], '%Y-%m-%d') + timedelta(days=random.randint(0,2))).strftime('%Y-%m-%d')
            expanded_rows.append(row)
        for _ in range(2):
            row = p.copy()
            row['Admission Type'] = ''
            row['Radiology'] = ''
            row['Laboratory'] = random.choice(laboratory_tests)
            row['Discharge Date'] = (datetime.strptime(p['Date of Admission'], '%Y-%m-%d') + timedelta(days=random.randint(0,2))).strftime('%Y-%m-%d')
            expanded_rows.append(row)

# ========== 3. Create final DataFrame and save ==========
df_final = pd.DataFrame(expanded_rows)
# Reorder columns as requested
column_order = ['Patient Name', 'Age', 'Gender', 'Blood Type', 'Medical Condition', 'Date of Admission',
                'Doctor', 'Department', 'Insurance Provider', 'Patient Type', 'Billing Amount', 'Room Number',
                'Admission Type', 'Discharge Date', 'Medication', 'Test Results', 'Radiology', 'Laboratory']
df_final = df_final[column_order]

# Save to CSV
df_final.to_csv('synthetic_healthcare_data.csv', index=False)
print(f"Saved {len(df_final)} rows to synthetic_healthcare_data.csv")
```

```python
from scipy import stats

# Separate billing amounts by gender
billing_male = df_final[df_final['Gender'] == 'Male']['Billing Amount']
billing_female = df_final[df_final['Gender'] == 'Female']['Billing Amount']

# Perform independent samples t-test
t_statistic, p_value = stats.ttest_ind(billing_male, billing_female)

print(f"T-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpret the results
alpha = 0.05 # Significance level
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. There is a statistically significant difference in Billing Amount between Male and Female patients.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no statistically significant difference in Billing Amount between Male and Female patients.")
```

```python
from scipy import stats

# Get unique medical conditions
medical_conditions = df_final['Medical Condition'].unique()

# Create a list of billing amounts for each medical condition
billing_by_condition = [df_final[df_final['Medical Condition'] == condition]['Billing Amount'] for condition in medical_conditions]

# Perform one-way ANOVA test
f_statistic, p_value = stats.f_oneway(*billing_by_condition)

print(f"F-statistic: {f_statistic:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpret the results
alpha = 0.05 # Significance level
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. There is a statistically significant difference in Billing Amount among different Medical Conditions.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no statistically significant difference in Billing Amount among different Medical Conditions.")
```
